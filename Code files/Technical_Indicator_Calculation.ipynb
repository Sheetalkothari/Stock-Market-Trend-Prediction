```python
import pandas as pd
import talib
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.dates import YearLocator
import datetime as dt
import pandas_datareader as web
import matplotlib.dates as mdates
from mplfinance.original_flavor import candlestick_ohlc
import mplfinance as mplfinance
```

```python


current_dir = os.getcwd()

data_file_path = os.path.join(current_dir, 'Data files', 'NFLX.csv')
output_file_path = os.path.join(current_dir, 'Data files', 'NFLX_Updated.csv')

df = pd.read_csv(data_file_path)

# Calculate SMA (Simple Moving Average)
sma = talib.SMA(df['Close'], timeperiod=14)
df['SMA'] = sma

# Calculate EMA (Exponential Moving Average)
ema = talib.EMA(df['Close'], timeperiod=14)
df['EMA'] = ema

# Calculate MACD (Moving Average Convergence Divergence)
macd, macd_signal, _ = talib.MACD(df['Close'])
df['MACD'] = macd
df['MACD_Signal'] = macd_signal

# Calculate RSI (Relative Strength Index)
rsi = talib.RSI(df['Close'])
df['RSI'] = rsi

# Calculate ATR (Average True Range)
atr = talib.ATR(df['High'], df['Low'], df['Close'])
df['ATR'] = atr

# Calculate OBV (On Balance Volume)
obv = talib.OBV(df['Close'], df['Volume'])
df['OBV'] = obv

# Calculate Percent R
percent_r = talib.WILLR(df['High'], df['Low'], df['Close'])
df['Percent_R'] = percent_r

# Calculate ADX (Average Directional Index)
adx = talib.ADX(df['High'], df['Low'], df['Close'], timeperiod=14)
df['ADX'] = adx

# Calculate CCI (Commodity Channel Index)
cci = talib.CCI(df['High'], df['Low'], df['Close'], timeperiod=20)
df['CCI'] = cci

# Calculate ROC (Rate of Change)
roc = talib.ROC(df['Close'], timeperiod=1)  # 1-period ROC is equivalent to %Change
df['ROC'] = roc

# Calculate %Change
percent_change = df['Close'].pct_change() * 100
df['Percent_Change'] = percent_change

# Calculate Signal Line (EMA)
signal_line = talib.EMA(df['Close'], timeperiod=9)
df['Signal_Line'] = signal_line

# Calculate %D and %K (Stochastic Oscillator)
slowk, slowd = talib.STOCH(df['High'], df['Low'], df['Close'], fastk_period=14, slowk_period=3, slowd_period=3)
df['%D'] = slowd
df['%K'] = slowk

# Save the updated dataset to the file
df.to_csv(output_file_path, index=False)
```

```python
current_dir = os.getcwd()
data_file_path = os.path.join(current_dir, 'Data files', 'Historic_Dataset.csv')

# Read the data into a DataFrame
df = pd.read_csv(data_file_path)
```

```python

for column in df.columns:
    # Iterate over each row in the column
    for i in range(len(df)):
        # Check if the value is null
        if pd.isnull(df.at[i, column]):
            # Find the next available value in the column
            j = i + 1
            while j < len(df) and pd.isnull(df.at[j, column]):
                j += 1
            # If a next value exists, copy it to the null value
            if j < len(df):
                df.at[i, column] = df.at[j, column]

# Write the updated data to a new CSV file
df.to_csv('Historic_Dataset.csv', index=False)
```

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Generate the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

df_indexed = df.set_index('Date')

# Group the data by the company name
grouped_data = df_indexed.groupby('Company Name')

# Plot each company's stock data with a different color
plt.figure(figsize=(12, 7))
colors = plt.cm.rainbow(np.linspace(0, 1, len(grouped_data)))
for i, (company, data) in enumerate(grouped_data):
    plt.plot(data['Close'], color=colors[i], label=company)

# Set the x-axis ticks to show every year
years = YearLocator()  # Place a tick every year
ax = plt.gca()  # Get the current axes
ax.xaxis.set_major_locator(years)



# Add labels, title, legend, and grid
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Stock Close Price Over Time')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
